# BUBBLE SORT

```python
def Bubble_Sort(item):
  print(id(item))
  n = len(item)
  for i in range(n-1):
    for j in range(1,n-i):
      if(item[j] < item[j-1]):
        item[j-1],item[j] = item[j],item[j-1]
    print(item)
    print('+'*100)
  return item
```

```python
#Unsorted_Data = [i for i in input("Enter Data with space").split(' ')]
Unsorted_Data = [10,9,8,7,6,5]
print(id(Unsorted_Data))
#Unsorted_Data = [10,9,8,5,6,7]
#Unsorted_Data = [5,6,7,8,9,10]

Bubble_Sort(Unsorted_Data)
```

# Selection Sort

```python
def Selection_Sort(item):
  n = len(item)
  print(id(item))
  for i in range(n):
    min_index = i
    for j in range(i,n):
      if(item[j] < item[min_index]):
        min_index = j
    item[i],item[min_index] = item[min_index],item[i]
    print(item)
    print('+'*100)
```

```python
#Unsorted_Data = [i for i in input("Enter Data with space").split(' ')]
Unsorted_Data = [10,9,8,7,6,5]
print(id(Unsorted_Data))
#Unsorted_Data = [10,9,8,5,6,7]
#Unsorted_Data = [5,6,7,8,9,10]
Selection_Sort(Unsorted_Data)
```

# Insertion Sort

```python
def Insertion_Sort(item):
 # print(id(item))
  for i in range(1, len(item)):
    key = item[i]
    j = i-1
    while j >= 0 and key < item[j] :
      item[j + 1] = item[j] 
      j -= 1
    item[j + 1] = key 
```
